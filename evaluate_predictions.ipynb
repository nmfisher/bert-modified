```python
import csv
import json
from sklearn.metrics import f1_score

predictions = json.load(open("/virtualmachines/models/bert_swagger/predictions.json","r"))
source = {}
correct = 0
incorrect = 0
blank = 0
with open("/home/hydroxide/projects/swagger-data-synthesizer/out/commands_test.csv","r") as infile:
    testreader = csv.reader(infile, delimiter=',')
    for entry in testreader:
        example_id = entry[0].replace('"',"")
        answer_text = entry[3].replace('"',"")
        if example_id not in predictions:
            print("Skipping")
            continue
        if predictions[example_id] == "":
            blank += 1
        if answer_text == predictions[example_id]:
            correct += 1
        else:
            print(entry)
            print(predictions[example_id])
            incorrect += 1
print("Accuracy : %s " % (correct / (correct + incorrect)))
print("Blank : %s " % (blank / (correct + incorrect)))
```
